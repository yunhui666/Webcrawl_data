```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import unquote

domain = 'https://www.shakeshack.com'
req = requests.get(f'{domain}/locations/')
soup = BeautifulSoup(req.content)
```

```python
stores = soup.find_all('div',{'class':'geolocation-location'})
stores
```

```python
store = stores[0]
lat = store.get('data-lat')
lng = store.get('data-lng')
link = store.find_all('a',{'class':'locations-map--pin--links--item'})[1]
address = unquote(link.get('href').split('=')[1].replace(',,',','))
address
```

```python
ny_location = []
for store in stores:
  lat = store.get('data-lat')
  lng = store.get('data-lng')
  link = store.find_all('a',{'class':'locations-map--pin--links--item'})[1]
  address = unquote(link.get('href').split('=')[1].replace(',,',','))
  if address.split(',')[-2] == 'NY':
    ny_location.append((lat,lng,address))
```

```python
import pandas as pd
addresses = pd.DataFrame(ny_location,columns = ['lat','lng','address'])
addresses
```

```python
addresses.to_csv('shakeshack_address.csv', index=None)
```
